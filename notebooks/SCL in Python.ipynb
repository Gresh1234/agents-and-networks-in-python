```python
# We set the matplotlib backend to ipympl so we can interact with plots, and import a few libraries.
%matplotlib ipympl
import matplotlib.pyplot as plt
from pyrosm import OSM
from pyrosm import get_data

# We open our trimmed Santiago Metropolitan Region PBF file and check if pyrosm loaded it successfully.
osm = OSM("../data/external/OSM/chile-rm-latest.osm.pbf")
print("Type of 'osm' instance: ", type(osm))
```

```python
# We load a few networks and buildings from the PBF file.
walk_net = osm.get_network(network_type="walking")
bike_net = osm.get_network(network_type="cycling")
drive_net = osm.get_network(network_type="driving")
buildings = osm.get_buildings()
```

```python
# We plot these networks all together to have a closer look at the data contained in the PBF.
ax = walk_net.plot(figsize=(6,6), color="brown")
ax = bike_net.plot(ax=ax, color="blue")
ax = drive_net.plot(ax=ax, color="gray")
ax = buildings.plot(ax=ax, facecolor="gray", edgecolor="black")
plt.show()
```
